```python
freq ={}
for value in df['parental level of education'].unique():
  c=0
  for item in df['parental level of education']:
    if item==value:
       c+=1

  freq[value]=c


plt.figure(figsize=(20,7))

plt.bar(range(len(df['parental level of education'].unique())),list(freq.values()))

plt.xticks(range(len(freq)),df['parental level of education'].unique())

plt.show()
```
